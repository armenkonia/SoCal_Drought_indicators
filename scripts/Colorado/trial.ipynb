```python
import pdfplumber
from PIL import Image, ImageDraw

# doc_path = r"..\Main\Bulletin 132-19 020323.pdf"
doc_path = r"C:\Users\armen\Desktop\Drought Indicators - SoCal\Main\Colorado\Extra\Water Accounting Reports\2023.pdf"
page_number = 63


with pdfplumber.open(doc_path) as pdf:
        page = pdf.pages[page_number - 1]  # Page numbers are 0-indexed in pdfplumber
        tables = page.extract_tables()
        im = page.to_image()
```

```python
draw = ImageDraw.Draw(im.original)  # Draw on the original image
line_x = 100  # X-coordinate where you want to draw the vertical line
line_width = 2  # Width of the line
draw.line([(line_x, 0), (line_x, im.original.height)], fill="red", width=line_width)
im.original.show()
```
